## Preparing the dataset

```python
dataset=pd.DataFrame(california.data, columns=california.feature_names)
```

```python
dataset.head()
```

```python
dataset['Price']=california.target
```

```python
dataset.head()
```

```python
dataset.info()
```

## Summarizing the stats of this data

```python
dataset.describe()
```

## Exploratory Data Analysis (EDA)

```python
## Check for missing values
dataset.isnull().sum()
```

```python
## Correlation

dataset.corr()
```

```python
import seaborn as sns
sns.pairplot(dataset)
```

```python
plt.scatter(dataset['MedInc'],dataset['Price'])
plt.xlabel('Median Income')
plt.ylabel('Price')
```

```python
plt.scatter(dataset['AveBedrms'],dataset['Price'])
plt.xlabel('Average Bedrooms')
plt.ylabel('Price')
plt.xlim(0, 10)
```

```python
sns.regplot(x='AveBedrms',y='Price', data=dataset)
```

```python
sns.regplot(x='Latitude',y='Price', data=dataset)
```

```python
## Independent and Dependent Features

X=dataset.iloc[:,:-1]
y=dataset.iloc[:,-1]

```

```python
X.head()
```

```python
y.head()
```

```python
## Train Test Split

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
```

```python
X_train=scaler.fit_transform(X_train)
```

```python
X_test=scaler.transform(X_test)
```

## Model Training

```python
from sklearn.linear_model import LinearRegression
```

```python
regression=LinearRegression()
```

```python
regression.fit(X_train,y_train)
```

```python
## Print the coefficients and the intercept
print(regression.coef_)
```

```python
## Which parameteres have been used to train the model?
regression.get_params()
```

```python
## Prediction With Test Data
reg_pred=regression.predict(X_test)
```

```python
reg_pred
```

```python
## plot a scatter plot for the prediction
plt.scatter(y_test,reg_pred)
```

```python
## Residuals
residuals=y_test-reg_pred
```

```python
residuals
```

```python
## Plot this residuals

sns.displot(residuals, kind="kde")
```

```python
  ## scater plot respect to the predictions and residuals
plt.scatter(reg_pred,residuals)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

print(mean_absolute_error(y_test,reg_pred))
print(mean_squared_error(y_test,reg_pred))
print(np.sqrt(mean_squared_error(y_test,reg_pred)))

```

# R square and adjusted R square

```python
from sklearn.metrics import r2_score
score=r2_score(y_test,reg_pred)
print(score)
```

```python
1 - (1-score)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)
```
